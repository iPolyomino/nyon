#### 具体例

f(0), f(1),f(5)を求めよ．また，f(x)=0となるxを求めよ．

```python
from sympy import *
def f(x):
    return 3*x-4
    
x = Symbol('x')
print(f(0))
print(f(1))
print(f(5))

solve(f(x),x)
```

#### 具体例2

y=f(x)とy=g(x)の２直線の交点を求めよ．さらに連立方程式を立てて求めよ．交点のx座標をy=f(x)-g(x)としてプロットし，確認せよ．

```python
%matplotlib inline
from sympy import *
def f(x):
    return 3*x-4
def g(x):
    return -2*x+4

x = Symbol('x')
p = plot(f(x),g(x),(x, -1, 3),legend=True, show=False)
p[0].line_color = 'b'
p[1].line_color = 'r'
p.show()
```

```python
solve(f(x)-g(x),x)
```

```python
x, y = symbols('x y')
solve({y-f(x),y-g(x)},(x,y))
```

```python
plot(f(x)-g(x),(x, -1, 3))
print(8/5)
```

### 逆関数

#### 逆関数の求め方
1. 定義域xと値域yを入れ替えx=f(y)という方程式をたてる．
2. これをyについて解く．
3. 得られた関数y=g(x)がy=f(x)の逆関数である．
  
２つの関数y=g(x)とy=f(x)とを同時にプロットすると 直線y=xに関して線対称の位置をとることがわかる．

#### 具体例

指数関数(exp)の逆関数を求め，プロットして，直線y=xに関して線対称であることを確かめよ．

```python
%matplotlib inline
from sympy import *

x, y = symbols('x y')
solve(x - exp(y),y)
```

```python
%matplotlib inline
from sympy import *

plot(exp(x),log(x),x,(x, -4,4), ylim=[-4,4])
```

### 単値関数と多値関数

#### 具体例

sin関数についても逆関数を求め，プロットして，直線y=xに関して線対称であるとともに，多値関数となっていることを確かめよ．

```python
%matplotlib inline
from sympy import *

x, y = symbols('x y')
solve(x - sin(y),y)
```

```python
%matplotlib inline
from sympy import *

p = plot(sin(x),-asin(x)+pi,asin(x),x,(x, -6,6), ylim=[-6,6],
         legend=True, show=False)
p[0].line_color = 'b'
p[1].line_color = 'r'
p[2].line_color = 'g'
p[3].line_color = 'm'
p.show()
```
