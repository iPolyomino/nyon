## 微分(Diff)-I

### 単純な微分

```python
from sympy import *
from sympy.interactive import printing
printing.init_printing(use_latex=True)
from sympy.plotting import plot
from IPython.display import display # display(code)で数式をキレイに表示可能

x,y = symbols('x y')

diff(x**2 - 3*x + 2, x)
```

```python
diff(sin(x),x,x)
```

```python
diff(x**4,x,3)
```

### 偏微分(PartialDiff)

```python
eq1=(x+y)/(x*y)
diff(eq1,x)
```

### 例題:関数の微分と増減表

```python
from IPython.display import display
init_printing(use_latex=True)

x = symbols('x')

f = x/(x**2-2*x+4)
df = diff(f,x)
display(f)
display(df)

p = plot(f,df,show=False)
p[0].line_color = 'b'
p[1].line_color = 'r'
p.show()
```

### 例題:接線

```python
x,y = symbols('x y')

f = x**3 - 2*x**2 - 35*x
df = diff(f,x)
display(f,df)

x0 = 3

"""
y-f(x) = df(x) * (x-a)
y = df(a) * (x-a) + f(a)
"""

f1 = df.subs({x:x0})*(x-x0) + f.subs({x:x0})
display(f1)
p = plot(f,f1,show=False)
p[0].line_color = 'b'
p[1].line_color = 'r'

p.show()
```

### 極限(limit)

```python
from sympy import *
x = symbols('x')
display(limit(x,x,0))
display(limit(x,x,oo))
display(limit(1/x,x,oo))
display(limit(x/sin(x),x,0))
```

```python
d = symbols('d')
f = x**2
df = diff(f,x)
display(df.subs({x:2}))

limit((f.subs({x:x+d}) - f.subs({x:x})) / d, d,0)
```

## 微分(Diff)-II

### 級数展開(series)

```python
from sympy import *
from sympy.plotting import plot
init_printing(use_latex=True)

x,a = symbols('x a')

sin(x).series(x,a,4)
```

```python
fs = series(sin(x),x,a,4)

f0 = fs.removeO()
display(f0)
f1 = f0.subs({a:pi})
display(f1)
```

```python
y = symbols('y')
p = plot(sin(x),f1, (x,0,2*pi), ylim=[-1,1], show=False)
p[0].line_color = 'b'
p[1].line_color = 'r'
p.show()
```

### 全微分

```python
from sympy import *

def total_derivatives(x,y):
    f = lambda x,y : x**4 * exp(-y**2)
    xdx = diff(f(x,y),x)
    ydx = diff(f(x,y),y)
    return xdx + ydx

x,y = symbols('x y')
total_derivatives(x,y)
```

### 複合関数の微分

```python
from sympy import *

f = lambda x : exp(x)
g = lambda x : cos(x)
diff(f(x)*g(x), x)
diff(f(g(x)), x)
```
