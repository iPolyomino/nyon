## $$\sin^2x , \cos^2x$$

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x)**2
def g(x):
    return np.cos(x)**2
```

$f(x) = \sin^2x, g(x) = \cos^2x$ を同時にプロット

```python
%matplotlib notebook
X = np.linspace(0, np.pi, 256, endpoint=True)
plt.plot(X, f(X))
plt.plot(X, g(X))
plt.show()
```

```python
%matplotlib notebook
X = np.linspace(0, np.pi, 256, endpoint=True)
plt.plot(X, f(X))
plt.plot(X, g(X))
plt.plot(X, f(X)+g(X))
plt.show()
```

$f(x)+g(x) = 1$ を示した。

### 恒等式

#### 定義
含まれている文字にどのような値を代入しても，その等式の両辺の値が存在する限り常に成り立つ等式

$$
\sin x^2 + \cos x^2 = 1 \\
f(x) = \sin x^2 \\
g(x)= \cos x^2
$$
より、$ f(x) + g(x) = 1 $ は恒等式である。

#### 参考
[§１　数　と　式](https://www.kwansei.ac.jp/hs/z90010/sugakua/kotosiki/kotosiki.htm)

### 3次元プロットから

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(0, np.pi, 256, endpoint=True)
plt.plot(X, np.cos(X))
plt.plot(X, np.cos(X+np.pi/2))
plt.plot(X, np.cos(X+np.pi))


plt.show()
```

```python
%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

def g(x,y):
    return np.cos(x+y)

x = np.arange(0, np.pi, 0.25)
y = np.arange(0, np.pi, 0.25)
X, Y = np.meshgrid(x, y)
Z1 = g(X,Y)

fig = plt.figure()
plot3d = Axes3D(fig)
plot3d.plot_surface(X,Y,Z1) 

plt.show()
```

```python
%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

def h(x,y):
    return np.cos(x)*np.cos(y)-np.sin(x)*np.sin(y)

x = np.arange(0, np.pi, 0.25)
y = np.arange(0, np.pi, 0.25)
X, Y = np.meshgrid(x, y)
Z1 = h(X,Y)

fig = plt.figure()
plot3d = Axes3D(fig)
plot3d.plot_surface(X,Y,Z1) 

plt.show()
```

それぞれ、
$$
f(x;c) = cos(x+c) \\
g(x,y) = cos(x+y) \\
h(x,y) = cos(x)cos(y) - sin(x)sin(y) \\
$$

g(x)をf(x)に一致させる

```python
%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return np.cos(2*x)
def g(x):
    return np.sin(x) ** 2


X = np.linspace(0, np.pi, 256, endpoint=True)
plt.plot(X, f(X))
plt.plot(X, g((X+np.pi/2))*2 - 1)

plt.show()
```

### 手順
1. 位相を $\frac{\pi}{2}$ ずらす
2. 振幅を2倍にする
3. $y$の値を 1 減らす

### $f(x)$ を $g(x)$ の項で表現する

```python
%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

def g(x):
    return np.sin(x) ** 2
def f(x):
    return (-g(x))*2 + 1

X = np.linspace(0, np.pi, 256, endpoint=True)
plt.plot(X, f(X))
plt.plot(X, g(X))

plt.show()
```

### 手順

$$
f(x) = \cos (2x)
$$
$$
g(x) = \sin^2 (x)
$$
のとき、

$$
f(x) = g((x+\frac{\pi}{2}))\times2 - 1
$$
